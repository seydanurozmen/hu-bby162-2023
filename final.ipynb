```python
from google.colab import drive
import os

drive.mount("/gdrive")
dosya_adi = "/gdrive/My Drive/veri.txt"

def eserleri_listele():
    try:
        yapilacak = open(dosya_adi, "r")
        eserler = yapilacak.readlines()
        yapilacak.close()

        if len(eserler) == 0:
            print("Katalog boş.")
        else:
            for eser in eserler:
                eser = eser.strip().split(" | ")
                if len(eser) == 3:
                    print("Yazar adı:", eser[0])
                    print("Kitap Adı:", eser[1])
                    print("Yayın tarihi:", eser[2])
                    print("--------------------------")
                else:
                    print("Eser verisi hatalı: ", eser)
    except FileNotFoundError:
        print("Katalog bulunamadı.")

def eser_ekle():
    try:
        Yazar = input("Yazar adı: ")
        KitapAdi = input("Kitap Adı: ")
        YayinTarihi = input("Yayın tarihi: ")

        with open(dosya_adi, "a") as dosya:
            dosya.write(Yazar + " | " + KitapAdi + " | " + YayinTarihi + "\n")
        print("Eser başarıyla eklendi.")
    except Exception as e:
        print("Eser eklenirken bir hata oluştu:", str(e))

def eser_sil():
    try:
        eser_adi = input("Silmek istediğiniz eser adını girin: ")

        if os.path.exists(dosya_adi):
            with open(dosya_adi, "r") as f:
                lines = f.readlines()

            if len(lines) == 0:
                print("Katalog boş.")
                return

            silindi = False
            with open(dosya_adi, "w") as f:
                for line in lines:
                    eser_bilgileri = line.strip().split(" | ")
                    if len(eser_bilgileri) == 3 and eser_bilgileri[1] != eser_adi:
                        f.write(line)
                    elif len(eser_bilgileri) == 3 and eser_bilgileri[1] == eser_adi:
                        silindi = True

            if silindi:
                print("Eser başarıyla silindi.")
            else:
                print("Belirtilen eser bulunamadı.")
        else:
            print("Katalog bulunamadı.")
    except Exception as e:
        print("Eser silinirken bir hata oluştu:", str(e))

def eser_guncelle():
    try:
        eser_adi = input("Güncellemek istediğiniz eser adını girin: ")

        if os.path.exists(dosya_adi):
            with open(dosya_adi, "r") as f:
                lines = f.readlines()

            if len(lines) == 0:
                print("Katalog boş.")
                return

            guncellendi = False
            with open(dosya_adi, "w") as f:
                for line in lines:
                    eser_bilgileri = line.strip().split(" | ")
                    if len(eser_bilgileri) == 3 and eser_bilgileri[1] != eser_adi:
                        f.write(line)
                    elif len(eser_bilgileri) == 3 and eser_bilgileri[1] == eser_adi:
                        yeni_yazar = input("Yeni yazar adını girin: ")
                        yeni_kitap = input("Yeni kitap adını girin: ")
                        yeni_yayin_tarihi = input("Yeni yayın tarihini girin: ")
                        f.write(yeni_yazar + " | " + yeni_kitap + " | " + yeni_yayin_tarihi + "\n")
                        guncellendi = True

            if guncellendi:
                print("Eser başarıyla güncellendi.")
            else:
                print("Belirtilen eser bulunamadı.")
        else:
            print("Katalog bulunamadı.")
    except Exception as e:
        print("Eser güncellenirken bir hata oluştu:", str(e))

def katalog_temizle():
    try:
        if os.path.exists(dosya_adi):
            open(dosya_adi, "w").close()
            print("Katalog başarıyla temizlendi.")
        else:
            print("Katalog bulunamadı.")
    except Exception as e:
        print("Katalog temizlenirken bir hata oluştu:", str(e))

def menu_goster():
    print("\nKatalog Uygulaması")
    print("Bütün eserleri listelemek için '1' tuşuna basınız")
    print("Eser eklemek için '2' tuşuna basınız")
    print("Seçtiğiniz eseri silmek için '3' tuşuna basınız")
    print("Seçtiğiniz eseri güncellemek için '4' tuşuna basınız")
    print("Kataloğu temizlemek için '5' tuşuna basınız")

while True:
    menu_goster()
    islem = input("İşlem seçiniz: ")

    if islem == "1":
        eserleri_listele()

    elif islem == "2":
        eser_ekle()

    elif islem == "3":
        eser_sil()

    elif islem == "4":
        eser_guncelle()

    elif islem == "5":
        katalog_temizle()

    else:
        print("Hatalı işlem seçildi. Lütfen tekrar deneyin.")
```
